Answer the business question

We compared the cost-effectiveness of service providers with and without sustainable practices using the Student's t-test on the wedding database. Because the p-value (9.952837709511204e-05) was less than a significance level of 0.05. Hence, the null hypothesis was rejected, indicating a statistically significant difference in cost-effectiveness. The findings show that in our sample, vendors who practice sustainable practices have a significant cost-effectiveness advantage. This implies that investing in sustainability may benefit vendors' cost-effectiveness.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#All Vendor sheet

file = "/Users/vivi/Desktop/all.xlsx"
wedding = pd.read_excel(io = file,
                         sheet_name = 0,
                         header = 0)
wedding.describe().round(decimals = 2)
```

```python
from scipy.stats import ttest_ind

# Divide wedding database into two categories: vendors who use sustainable practices and those who don't.
sustainable_group = wedding[wedding['vendor_sustainable'] == 1]['price_ce']
non_sustainable_group = wedding[wedding['vendor_sustainable'] == 0]['price_ce']

# ttest
t_statistic, p_value = ttest_ind(sustainable_group, non_sustainable_group)

# result 
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Significance level of 0.05
alpha = 0.05
if p_value < alpha:
    print("""Cost-effectiveness differences among vendors with sustainable practices were 
found to be statistically significant.""")
else:
    print("""Cost-effectiveness differences observed among vendors 
with sustainable practices are not statistically significant.""")
```
